```sql

---------------------------------------------------------------------------------
-- Exploring columns on both tables 
---------------------------------------------------------------------------------
SELECT *
FROM workspace.default.userprofiles
LIMIT 10;

SELECT *
FROM workspace.default.viewership
LIMIT 10;

----------------------------------------------------------------------------------
--- Checking Date range
----------------------------------------------------------------------------------

-- They started collecting data on  2016-01-01
SELECT MIN(RecordDate2) AS  Min_date
FROM workspace.default.viewership;

-- They last collected data on 2016-03-31
SELECT MAX(RecordDate2) AS Max_date
FROM workspace.default.viewership;

-- The duration on the data is 3 months

----------------------------------------------------------------------------------
-- Checking the names of the different shows broadcasted 
-----------------------------------------------------------------------------------

-- We have 21 shows that are currently being broadcasted
SELECT DISTINCT Channel2
FROM workspace.default.viewership;

SELECT COUNT(DISTINCT Channel2) AS Number_of_Shows
FROM workspace.default.viewership;

------------------------------------------------------------------------------------
-- Joining 2 Tables userviewship table and userprofiles table
----------------------------------------------------------------------------------

SELECT
    -- User Profile columns
    up.UserID        AS up_UserID,
    up.Gender,
    up.Race,
    up.Age,
    up.Province,

    -- Viewership columns
    v.UserID0         AS v_UserID,
    v.Channel2,
    v.RecordDate2,
    v.`Duration 2`

FROM workspace.default.userprofiles AS up
FULL OUTER JOIN workspace.default.viewership AS v
    ON up.UserID = v.UserID0;

SELECT
    -- User Profile columns
    up.UserID        AS up_UserID,
    up.Gender,
    up.Race,
    up.Age,
    up.Province,

    -- Viewership columns
    v.UserID0         AS v_UserID,
    v.Channel2,
    v.RecordDate2,
    v.`Duration 2`,

    -- Age Buckets
    CASE
        WHEN up.Age < 12                    THEN 'Kid'
        WHEN up.Age BETWEEN 13 AND 19       THEN 'Teenager'
        WHEN up.Age BETWEEN 20 AND 34       THEN 'Young Adult'
        WHEN up.Age BETWEEN 35 AND 59       THEN 'Matured Adult'
        WHEN up.Age >= 60                   THEN 'Senior Citizen'
        ELSE 'Unknown'
    END AS Age_Group,

    -- Day Buckets
    CASE
        WHEN DAYOFWEEK(v.RecordDate2) IN (1, 7) THEN 'Weekend'   -- 1 = Sunday, 7 = Saturday
        ELSE                                          'Weekday'
    END AS Day_Type,

     -- Day of the Week Bucket
    CASE DAYOFWEEK(v.RecordDate2)
        WHEN 1 THEN 'Sunday'
        WHEN 2 THEN 'Monday'
        WHEN 3 THEN 'Tuesday'
        WHEN 4 THEN 'Wednesday'
        WHEN 5 THEN 'Thursday'
        WHEN 6 THEN 'Friday'
        WHEN 7 THEN 'Saturday'
        ELSE 'Unknown'
    END AS Day_Of_Week,

    -- Time Buckets
    CASE
        WHEN HOUR(v.RecordDate2) BETWEEN 5  AND 11  THEN 'Morning'    -- 05:00 - 11:59
        WHEN HOUR(v.RecordDate2) BETWEEN 12 AND 16  THEN 'Afternoon'  -- 12:00 - 16:59
        WHEN HOUR(v.RecordDate2) BETWEEN 17 AND 20  THEN 'Evening'    -- 17:00 - 20:59
        WHEN HOUR(v.RecordDate2) BETWEEN 21 AND 23  THEN 'Night'      -- 21:00 - 23:59
        WHEN HOUR(v.RecordDate2) BETWEEN 0  AND 4   THEN 'Midnight'   -- 00:00 - 04:59
        ELSE 'Unknown'
    END AS Time_Of_Day

FROM workspace.default.userprofiles AS up
FULL OUTER JOIN workspace.default.viewership AS v
    ON up.UserID = v.UserID0;
    
----------------------------------------------------------------------------------------
-- Clean up null values in the dataset 
----------------------------------------------------------------------------------------

SELECT 
    IFNULL(ROUND(UserID0, 0),'0') AS User_id, 
    IFNULL(Gender, 'No_Gender') AS Gender, 
    IFNULL(Race, 'No_race') AS Race,
    IFNULL(ROUND(Age, 0), '0') AS Age, 
    IFNULL(Province, 'No_Province') AS Province, 
    IFNULL(Channel2,'No_Channel') AS Channel2,
    IFNULL (RecordDate2, 'No_Recorddate') AS Recorddate2,
    TO_TIMESTAMP(Recorddate2, 'MM/DD/YYYY HH24:MI') AS Timestamp_value,
    CAST(To_timestamp(Recorddate2, 'MM/DD/YYYY HH24:MI') AS Date) Date_part,
    DATE_FORMAT(To_timestamp(Recorddate2, 'MM/DD/YYYY HH24:MI'), 'HH:mm:ss') AS Time_part

    FROM workspace.default.bright_tv;

    
```
